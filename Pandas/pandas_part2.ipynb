# Handling missing data - part 2

```python
var2 = pd.read_csv("C:\\Users\\vishe\\OneDrive\\Desktop\\Data Science\\Data_Science_Learning\\Pandas\\state-population.csv")
var2
```

```python
var2.dropna() # dropna() function drops the rows which have nan values
# axis -> using this parameter we can drop the row or columns having nan values for column axis =1 
# how -> using this parameter we can drop all the row having nan if how = "any" and if how = "all" then we can all drop that row which is having all values as nan 
```

```python
var2.dropna(how = "all") # -> eg
```

```python
var2.dropna(subset="population") # using subset parameter we can remove the nan values of specific column
```

```python
var2.dropna(thresh=4) # using thresh parameter we can remove the nan values upto the thresh value for eg if thresh = 3 then rows having nan values less than or equal to 3 will be drop
```

```python
var2.fillna("-") # fillna function is use to fill the nan value with the desire value
```

```python
var2.fillna({"state/region":"US","ages":"invalid","year":"-","population":"not_available"})
# if we want some specific value in a particular column then we have to pass it using dictionary
# method parameter is used to fill data in forward or backward direction
# inplace parameter is used to make a new data set from existing one
# limit parameter is used to make a certain nan values equals to the given string for eg "python",limit=1 then the column having 1 nan value will be fill with python
```
