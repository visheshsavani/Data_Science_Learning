# Join and Append function

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var2 = pd.DataFrame({"A":[1,2],"C":[21,22]},index=["a","b"])
var1.join(var2,how="left",lsuffix="_1",rsuffix="_2") # if 2 dataframe has same column name then lsuffix is required to avoid value error problem
# here how will work same as joins of sql
```

```python
var3 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var4 = pd.DataFrame({"D":[1,2],"C":[21,22]})
var4.append(var3) # now append function is not available so use concatenate funct to merge thats why error
```

# Pivot Table and Melt Functions

```python
x = pd.DataFrame({"days":[1,2,3,4,5,6],"eng":[18,16,19,12,17,18],"maths":[17,13,14,11,19,15]})
x.melt(id_vars="days",var_name="Subjects",value_name="Marks") # melt functions is use to change the shape of dataframe
# id_vars parameter is used to make a particular column id
# var_name parameter is used to give name to a variable column
# value_name parameter is used to give name to a value column
```

```python
# pivot is also used to reshape but according to some pattern
x = pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","b","c","a","c","b"],"eng":[18,16,19,12,17,18],"maths":[17,13,14,11,19,15]})
x.pivot_table(index="days",columns="st_name")
# we can also find statistical info like mean median and mode using aggfunc parameter
# there are also other parameters also to explore
```
