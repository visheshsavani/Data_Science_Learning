# Creating a CSV File

```python
import pandas as pd
dict = {'Comps':[200,16,23],'CSE':[21,12,98],'EXTC':[10,36,53],}
h = pd.DataFrame(dict)
#print(h)
#h.to_csv("Department_details.csv",index=False)
h.to_csv("Department_details.csv",index=False,header=['Marks1','Marks22','Marks3']) # in this way we can create and remove the index and change the header name
print(h)
```

# Arithmetic Operations and Logical Operations

```python
h["Total"] = h["Comps"]+h["CSE"]+h["EXTC"] # We can even do multiplication subtraction and division
print(h)
```

```python
h["pass/fail"] = h['CSE'] >=20 # Logical Operatiions
h
```

# Insert and Delete

```python
h.insert(3,"ECE",[22,11,9]) #Insert syntax(index,name,data) -> one method is this
```

```python
print(h)
```

```python
# another method to insert
h["div"] = ["A",'B','C']
```

```python
print(h)
```

```python
# in this method we can copy only data which is needed
h['k'] = h['Electronics'][:2] -> syntax
print(h)
```

```python
# for deletion there are two methods
#1st is using pop function
#2nd is using del keyword
h.pop('div')
h.pop('k')
```
