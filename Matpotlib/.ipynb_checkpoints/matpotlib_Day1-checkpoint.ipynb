# Matplotlib Concepts

# Bar Graph

```python
import matplotlib.pyplot as plt # Basics how to make and add color to bar graph
x = [1,2,3,4]
y = [2,4,6,8]
c = ["r","y","g","b"]
#plt.plot(x,y,color=c)
plt.bar(x,y,color=c)
plt.show()
```

```python
# Bar Graph parameters
import matplotlib.pyplot as plt
# All these are basic parameters
x1 = [1,2,3,4]
y1 = [2,4,6,8]
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of y = 2x")
plt.bar(x1,y1,width=0.4,color="m",edgecolor="r",align="edge")
plt.show()
```
